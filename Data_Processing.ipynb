```
# 数値ではない型の要素の抽出
pic = listings[['number_of_reviews']][listings['number_of_reviews'].apply(lambda s:pd.to_numeric(s, errors='coerce')).isnull()]
pic.index
listings = listings.drop(pic.index)
```

```
listings = listings.astype({"number_of_reviews":float})
```

```
# インポート
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()

# 関数例
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    return score

sentiment_analyzer_scores("I'm so bad!")
```

```
id_list = reviews["listing_id"].unique()
id_list.sort()
id_list
```

```
for i, id in enumerate(id_list):
    list_comments = reviews["comments"].loc[reviews["listing_id"]==id].values
    comment = ""
    for c in range(len(list_comments)):
        comment += str(list_comments[c])
    with open("/content/drive/My Drive/output_data/comment"+str(i+1)+".txt", 'w') as f:
        f.write(comment)
```
