<h1>Data Intagration</h1>
<p>各地域のlistings,reviewsのコメント数上位N個を抜粋し、listings.csv,reviews.csvとして格納。</p>

```
import pandas as pd
```

```
import glob
```

```
files = sorted(glob.glob("/content/drive/My Drive/input_data/listings/*.csv"))
listing_list = []
id_list = [] #抜粋した施設のidを格納。
print("listingデータの統合")
for i, path in enumerate(files):
    listing = pd.read_csv(path)
    listing_list.append(listing.sort_values('number_of_reviews',ascending=False)[:100]) #上位　50　個の施設を抜粋。
    id = listing.sort_values('number_of_reviews',ascending=False)[:100]["id"].values
    id_list.append(id)
    print(path,(i+1),"finished")
listings = pd.concat(listing_list)
for i in range(len(listing_list)):
    print((i+1),"番目のデータ：",listing_list[i].shape)
print("統合後：",listings.shape)
```

```
files = sorted(glob.glob("/content/drive/My Drive/input_data/reviews/*.csv"))
review_list = []
print("reviewデータの統合")
for i, path in enumerate(files):
    review = pd.read_csv(path)
    review_list.append(review.loc[review['listing_id'].isin(id_list[i])]) #id_listにある施設のレビューデータを抜粋。
    print(path,(i+1),"finished")
reviews = pd.concat(review_list)
for i in range(len(review_list)):
    print((i+1),"番目のデータ：",review_list[i].shape)
print("統合後：",reviews.shape)
```

```
listings.to_csv("/content/drive/My Drive/input_data/listings.csv",index=False)
reviews.to_csv("/content/drive/My Drive/input_data/reviews.csv",index=False)
```

```
loc_list = ["Chicago", "Miami", 'Hawaii', "Las Vegas", "Los an Angeles", "Nashvile", "New Orleans", "New York", "San Diego", "San Francisco"]

loc_id = pd.DataFrame({"id":[i for i in range(len(listings))],
                        "loc":[str(i) for i in range(len(listings))]})
i = 0
for j in range(len(loc_list)):
    for t in range(len(id_list[j])):
        loc_id['id'][i] = id_list[j][t]
        loc_id['loc'][i] = loc_list[j]
        i += 1
```
